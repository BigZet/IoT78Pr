```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
data = pd.read_csv('data.csv', sep = ";", header=0)


data['time'] = data['time'].map(lambda x : datetime.strptime(x, "%H:%M:%S"))

data
```

```python
def grap(p:str):
    t = plt.hist(data[p], bins = 3)
    plt.title(f"График распределения для {p}")
    plt.show()
    sns.lineplot(x=data['time'], y=data[p])
    plt.title(f"График динамики для {p}")
    plt.show()
    plt.pie(t[0], labels = [f"{np.round(t[1][i], 5)}-{np.round(t[1][i-1], 5)}" for i in range(1, len(t[1]))])
    plt.title(f"Круговая диаграмма для {p}")
    plt.show()
```

```python
for i in ('sound', 'illumination', 'voltage'):
    grap(i)
```
